```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading data
dataset = pd.read_csv('data (2).csv', header=None)
X = dataset[0]
y = dataset[1]
x_bar = np.mean(X)
y_bar = np.mean(y)
```

```python
# Part (a) - Normal equation
def line_of_best_fit(x,y, x_bar, y_bar):
    num = np.sum(x*y - y_bar*x)
    denom = np.sum(x**2 - x_bar*x)
    B = num/denom
    a = y_bar - B*x_bar 
    return a, B
```

```python
a, B = line_of_best_fit(X,y, x_bar, y_bar)
print('Value of a:', np.round(a,3))
print('Value of B:', np.round(B,3))

plt.scatter(X,y, c = 'red')
plt.plot(X, B*X + a, 'blue')
plt.show()
```

```python
# Part (b) - Gradient descent
def gradient_descent(x,y, lr):
    m,c = 0,0
    n = len(y)
    
    for i in range(100):
        y_pred = m*x + c
        dm = -2/n * np.sum(x*(y - y_pred))
        dc = -2/n * np.sum(y - y_pred)

        m = m - (lr*dm)
        c = c - (lr*dc)

    return m,c
```

```python
m,c = gradient_descent(X, y, lr = 0.001)
print("Value of m:", np.round(m,3))
print("Value of c:", np.round(c,3))

print("With learning rate 0.001, we see that the regression line doesn't fit the data well")
plt.scatter(X,y, c = 'red')
plt.plot(X, m*X + c, 'blue')
plt.show()
```

```python
m_2,c_2 = gradient_descent(X, y, lr = 0.0001)
print("Value of m:", np.round(m_2,3))
print("Value of c:", np.round(c_2,3))

print("With learning rate 0.0001, we see that the regression line fits the data well")
plt.scatter(X,y, c = 'red')
plt.plot(X, m_2*X + c_2, 'blue')
plt.show()
```
